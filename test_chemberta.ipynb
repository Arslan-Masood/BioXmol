```python
import torch
from transformers import AutoModel, AutoTokenizer
import numpy as np

model_name = "DeepChem/ChemBERTa-77M-MTR"

# Load model and tokenizer
print(f"Loading model: {model_name}")
model = AutoModel.from_pretrained(model_name, trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)

# Set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = model.to(device)
model.eval()
print(f"Model loaded on device: {device}")
```

```python
# Test with sample SMILES
smiles = ["Cn1c(=O)c2c(ncn2C)n(C)c1=O", "CC(=O)Oc1ccccc1C(=O)O", "C1=CC=CC=C1"]

print(f"Testing with {len(smiles)} SMILES:")
for s in smiles:
    print(f"  - {s}")

# Tokenize and get features
inputs = tokenizer(smiles, padding=True, return_tensors="pt")
inputs = {k: v.to(device) for k, v in inputs.items()}

print(f"\nInput shape: {inputs['input_ids'].shape}")

# Get features
with torch.no_grad():
    outputs = model(**inputs)
    features = outputs.pooler_output.cpu().numpy()

print(f"\nFeatures shape: {features.shape}")
print(f"Feature dimension: {features.shape[1]}")
print(f"\nFirst feature vector (first SMILES):")
print(features[0][:10], "...")  # Show first 10 values
```

```python
# Test with single SMILES
single_smiles = "CCO"  # Ethanol
inputs_single = tokenizer(single_smiles, padding=True, return_tensors="pt")
inputs_single = {k: v.to(device) for k, v in inputs_single.items()}

with torch.no_grad():
    outputs_single = model(**inputs_single)
    feature_single = outputs_single.pooler_output.cpu().numpy()

print(f"Single SMILES: {single_smiles}")
print(f"Feature shape: {feature_single.shape}")
print(f"Feature vector (first 10 values): {feature_single[0][:10]}...")
```
