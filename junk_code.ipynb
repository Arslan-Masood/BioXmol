```python
def split_dilirank_temporal(input_file: str,train_cutoff_year: int = 2010):
    """
    Split DILIRank dataset into train/test based on approval year.
    
    Args:
        input_file: Path to input CSV file (should have binary labels)
        output_dir: Directory to save train/test splits
        train_cutoff_year: Year threshold for train/test split (default: 2010)
    """
    # Load data
    df = pd.read_csv(input_file)
    print(f"Loaded {len(df)} compounds")
    
    # Check if ChEMBL_First_Approval column exists
    if 'ChEMBL_First_Approval' not in df.columns:
        raise ValueError("ChEMBL_First_Approval column not found in input file")
    
    # Check if binary_label column exists
    if 'binary_label' not in df.columns:
        raise ValueError("binary_label column not found. Please use binary_standard or binary_conservative file")
    
    # Remove rows with missing approval year
    df = df[df['ChEMBL_First_Approval'].notna()].copy()
    print(f"After filtering for non-null ChEMBL_First_Approval: {len(df)} compounds")
    
    # Split based on approval year
    train_df = df[df['ChEMBL_First_Approval'] < train_cutoff_year].copy()
    test_df = df[df['ChEMBL_First_Approval'] >= train_cutoff_year].copy()
    
    print(f"\nTemporal Split (cutoff year: {train_cutoff_year}):")
    print(f"  Train (< {train_cutoff_year}): {len(train_df)} compounds")
    print(f"  Test (>= {train_cutoff_year}): {len(test_df)} compounds")
    
    # Statistics
    if len(train_df) > 0:
        print(f"\nTrain set:")
        print(f"  DILI-Positive (1): {(train_df['binary_label'] == 1).sum()}")
        print(f"  DILI-Negative (0): {(train_df['binary_label'] == 0).sum()}")
        if (train_df['binary_label'] == 0).sum() > 0:
            print(f"  Class balance ratio: {(train_df['binary_label'] == 1).sum() / (train_df['binary_label'] == 0).sum():.2f}")
        print(f"  Approval year range: {train_df['ChEMBL_First_Approval'].min():.0f} - {train_df['ChEMBL_First_Approval'].max():.0f}")
    
    if len(test_df) > 0:
        print(f"\nTest set:")
        print(f"  DILI-Positive (1): {(test_df['binary_label'] == 1).sum()}")
        print(f"  DILI-Negative (0): {(test_df['binary_label'] == 0).sum()}")
        if (test_df['binary_label'] == 0).sum() > 0:
            print(f"  Class balance ratio: {(test_df['binary_label'] == 1).sum() / (test_df['binary_label'] == 0).sum():.2f}")
        print(f"  Approval year range: {test_df['ChEMBL_First_Approval'].min():.0f} - {test_df['ChEMBL_First_Approval'].max():.0f}")
```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
#from cuml import UMAP as cumlUMAP
import argparse
import warnings
import umap
warnings.filterwarnings('ignore')

import os
import numpy as np
import pandas as pd
import pyarrow.dataset as ds
import time
```

```python
def read_filtered_parquet(file_path, selected_drugs):
    """
    Efficiently reads parquet file and filters rows based on selected InChIKeys using pyarrow.dataset.
    """
    dataset = ds.dataset(file_path, format="parquet")
    table = dataset.to_table(filter=ds.field("Metadata_InChIKey").isin(selected_drugs))
    return table.to_pandas()

def load_aggregated_data_by_drugs(data_dir, max_drugs=10):
    """
    Load the aggregated parquet files and sample by drugs (including all replicates)
    """
    print("Loading aggregated JUMP CP data by drugs...")
    
    # Load both files
    original_file = os.path.join(data_dir, "filtered.parquet")
    centered_file = os.path.join(data_dir, "centered.filtered.parquet")
    
    # Step 1: Read only drug metadata to find common drugs
    print("Reading drug metadata to find common drugs...")
    orig_drugs = set(pd.read_parquet(original_file, columns=['Metadata_InChIKey'], engine="pyarrow")['Metadata_InChIKey'].dropna().unique())
    cent_drugs = set(pd.read_parquet(centered_file, columns=['Metadata_InChIKey'], engine="pyarrow")['Metadata_InChIKey'].dropna().unique())
    common_drugs = list(orig_drugs.intersection(cent_drugs))
    
    # Remove NaN/empty drugs if any
    common_drugs = [drug for drug in common_drugs if pd.notna(drug) and drug != '']
    
    print(f"Common drugs: {len(common_drugs)} (orig: {len(orig_drugs)}, cent: {len(cent_drugs)})")
    
    # Sample drugs
    if len(common_drugs) > max_drugs:
        selected_drugs = np.random.choice(common_drugs, size=max_drugs, replace=False)
    else:
        selected_drugs = common_drugs
    
    print(f"Using {len(selected_drugs)} drugs")
    
    # Now read only the data for selected drugs using filters
    print("Loading filtered data...")
    df_orig_filtered = read_filtered_parquet(original_file, selected_drugs)
    df_cent_filtered = read_filtered_parquet(centered_file, selected_drugs)
    
    print(f"Original: {df_orig_filtered.shape}, Centered: {df_cent_filtered.shape}")
    print(f"Original drugs: {df_orig_filtered['Metadata_InChIKey'].nunique()}, plates: {df_orig_filtered['Metadata_Plate'].nunique()}")
    print(f"Centered drugs: {df_cent_filtered['Metadata_InChIKey'].nunique()}, plates: {df_cent_filtered['Metadata_Plate'].nunique()}")
    
    return {
        'original': df_orig_filtered,
        'centered': df_cent_filtered
    }
```

```python
def prepare_features(df):
    """
    Extract feature columns for UMAP
    """
    # Get feature columns (non-metadata)
    feature_cols = [col for col in df.columns 
                   if not col.startswith('Metadata_')]
    
    print(f"Using {len(feature_cols)} feature columns")
    
    # Extract features
    features = df[feature_cols].copy()
    
    # Handle missing/infinite values
    features = features.replace([np.inf, -np.inf], np.nan)
    features = features.fillna(features.median())
    
    # Remove constant columns
    varying_cols = features.columns[features.nunique() > 1]
    features = features[varying_cols]
    
    print(f"Using {len(varying_cols)} varying features after filtering")
    
    return features.values, varying_cols.tolist()
```

```python
def create_umap_plots_by_drugs(data_dict, output_dir='umap_plots', n_neighbors=15, min_dist=0.1):
    """
    Create UMAP plots for both datasets colored by drugs
    """
    print(f"\n🧬 Starting Drug-based UMAP analysis at {time.strftime('%Y-%m-%d %H:%M:%S')}")
    analysis_start_time = time.time()
    
    os.makedirs(output_dir, exist_ok=True)
    
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle('UMAP: Original vs Normalized JUMP Cell Painting Data (Colored by Drug)', fontsize=16, fontweight='bold')
    
    results = {}
    total_embedding_time = 0
    
    for idx, (data_type, df) in enumerate(data_dict.items()):
        print(f"\nProcessing {data_type} data for Drug UMAP...")
        
        # Prepare features
        features, feature_names = prepare_features(df)
        
        if features.shape[0] == 0 or features.shape[1] == 0:
            print(f"No valid features for {data_type} data")
            continue
        
        # Run UMAP
        print("Computing UMAP embedding...")
        start_time = time.time()
        
        
        reducer = umap.UMAP(n_neighbors=n_neighbors, min_dist=min_dist, random_state=42, verbose=False)
        embedding = reducer.fit_transform(features)
        
        end_time = time.time()
        embedding_time = end_time - start_time
        
        print(f"UMAP embedding shape: {embedding.shape}")
        print(f"⏱️  UMAP computation time for {data_type} data: {embedding_time:.2f} seconds ({embedding_time/60:.2f} minutes)")
        
        total_embedding_time += embedding_time
        
        # Plot colored by drug
        ax = axes[idx]
        if 'Metadata_InChIKey' in df.columns and df['Metadata_InChIKey'].nunique() > 1:
            drug_ids = df['Metadata_InChIKey'].values
            le = LabelEncoder()
            colors = le.fit_transform(drug_ids)
            unique_drugs = le.classes_
            
            scatter = ax.scatter(embedding[:, 0], embedding[:, 1], 
                               c=colors, cmap='tab20', alpha=0.7, s=20)
            
            # Add legend if not too many drugs
            if len(unique_drugs) <= 15:
                for i, drug in enumerate(unique_drugs):
                    mask = colors == i
                    if np.any(mask):
                        # Truncate drug name for display
                        drug_label = str(drug)[:12] + "..." if len(str(drug)) > 12 else str(drug)
                        ax.scatter([], [], c=plt.cm.tab20(i), label=drug_label, s=50)
                ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=8, title='Drug ID')
            
            ax.set_title(f'{data_type.title()} Data - Colored by Drug\n({len(unique_drugs)} drugs, {df["Metadata_Plate"].nunique()} plates)')
        else:
            ax.scatter(embedding[:, 0], embedding[:, 1], alpha=0.7, s=20)
            ax.set_title(f'{data_type.title()} Data')
        
        ax.set_xlabel('UMAP 1')
        ax.set_ylabel('UMAP 2')
        ax.grid(True, alpha=0.3)
        
    
    plt.tight_layout()
    
    return fig
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def plot_measurement_distributions(siginfo_merged: pd.DataFrame, data_dir: str) -> None:
    """
    Plot distributions of doses, times, and replicates across compounds.
    
    Args:
        siginfo_merged: DataFrame with merged signature information
        data_dir: Directory to save the plots
    """
    # Calculate distributions
    dose_summary = (siginfo_merged.groupby('canonical_smiles')['pert_dose']
                   .nunique()
                   .value_counts()
                   .sort_index())
    
    time_summary = (siginfo_merged.groupby('canonical_smiles')['pert_time']
                   .nunique()
                   .value_counts()
                   .sort_index())
    
    replicate_summary = (siginfo_merged.groupby(['canonical_smiles', 'cell_iname', 
                                               'pert_dose', 'pert_time'])
                        .size()
                        .value_counts()
                        .sort_index())

    # Create subplots
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

    # Plot 1: Doses per compound
    sns.barplot(x=dose_summary.index, y=dose_summary.values, ax=ax1, color='skyblue')
    ax1.set_yscale('log')
    # Add value labels on top of bars
    for i, v in enumerate(dose_summary.values):
        ax1.text(i, v, str(v), ha='center', va='bottom')
    ax1.set_title('Number of Doses per Compound')
    ax1.set_xlabel('Number of Unique Doses')
    ax1.set_ylabel('Number of Compounds (log scale)')

    # Plot 2: Times per compound
    sns.barplot(x=time_summary.index, y=time_summary.values, ax=ax2, color='lightgreen')
    ax2.set_yscale('log')
    # Add value labels on top of bars
    for i, v in enumerate(time_summary.values):
        ax2.text(i, v, str(v), ha='center', va='bottom')
    ax2.set_title('Number of Time Points per Compound')
    ax2.set_xlabel('Number of Unique Time Points')
    ax2.set_ylabel('Number of Compounds (log scale)')

    # Plot 3: Replicates per condition
    sns.barplot(x=replicate_summary.index, y=replicate_summary.values, ax=ax3, color='salmon')
    ax3.set_yscale('log')
    # Add value labels on top of bars
    for i, v in enumerate(replicate_summary.values):
        ax3.text(i, v, str(v), ha='center', va='bottom')
    ax3.set_title('Number of Replicates per Condition')
    ax3.set_xlabel('Number of Replicates')
    ax3.set_ylabel('Number of Conditions (log scale)')

    plt.tight_layout()
    save_path = os.path.join(data_dir, 'measurement_distributions.png')
    plt.savefig(save_path, dpi=300, bbox_inches='tight')
    plt.close()

data = pd.read_parquet("/scratch/cs/pml/AI_drug/molecular_representation_learning/LINCS/landmark_cmp_data_min1000compounds_all_measurements.parquet")
plot_measurement_distributions(data, "/scratch/cs/pml/AI_drug/molecular_representation_learning/LINCS")
```

```python
# Parameters from your config
data_path = "/scratch/work/masooda1/mocop/data/jump_data/cell_fetures_with_smiles_1000.parquet"
genomic_data_path = "/scratch/cs/pml/AI_drug/molecular_representation_learning/LINCS/landmark_cmp_data_U2OS_single_dose_time.parquet"
pad_length = 250
batch_size = 64
num_workers = 0

# Create dataset
dataset = TripleInputGraphDatasetJUMP(
    data_path=data_path,
    genomic_data_path=genomic_data_path,
    pad_length=pad_length
)

# Create dataloader
dataloader = DataLoader(
    dataset,
    batch_size=batch_size,
    num_workers=num_workers,
    pin_memory=True,
    shuffle=True
)

# Test the dataloader
print(f"Dataset size: {len(dataset)}")
batch = next(iter(dataloader))

print("\nGraph inputs:")
adj_mat, node_feat, atom_vec = batch['inputs']['x_a'][0], batch['inputs']['x_a'][1], batch['inputs']['x_a'][2]
print(f"Adjacency matrix shape: {adj_mat.shape}")
print(f"Node features shape: {node_feat.shape}")
print(f"Atom vector shape: {atom_vec.shape}")

print("\nMorphological features:")
print(f"Shape: {batch['inputs']['x_b'].shape}")
# Count examples with missing morphological features
missing_morph = (batch['inputs']['x_b'][:, 0] == -1).sum().item()
print(f"Number of examples with missing morphological features: {missing_morph}/{batch_size}")

print("\nGenomic features:")
print(f"Shape: {batch['inputs']['x_c'].shape}")
# Count examples with missing genomic features
missing_genomic = (batch['inputs']['x_c'][:, 0] == -1).sum().item()
print(f"Number of examples with missing genomic features: {missing_genomic}/{batch_size}")

# Count examples missing both types of features
missing_both = ((batch['inputs']['x_b'][:, 0] == -1) & (batch['inputs']['x_c'][:, 0] == -1)).sum().item()
print(f"\nNumber of examples missing both types of features: {missing_both}/{batch_size}")

print("\nLabels:")
print(f"Shape: {batch['labels'].shape}")
```

```python
# Create model
model = TripleInputEncoder(
    encoder_a=GatedGraphNeuralNetwork(
        n_edge=1,
        in_dim=75,
        n_conv=6,
        fc_dims=[1024],
        p_dropout=0.1
    ),
    encoder_b=MultiLayerPerceptron(
        num_input_features=len(dataset.morph_cols),  # morphological features
        hidden_layer_dimensions=[512, 256, 128],
        p_dropout=0.1
    ),
    encoder_c=MultiLayerPerceptron(
        num_input_features=len(dataset.genomic_cols),  # genomic features
        hidden_layer_dimensions=[512, 256, 128],
        p_dropout=0.1
    ),
    supervised_head_dim=[202],
    non_lin_proj=True,
    dim=128,
    temperature=10
)

# Set optimizer
optimizer = torch.optim.AdamW(model.parameters(), lr=0.001)
model.set_optimizer(optimizer)

# Test forward pass
print("\nTesting forward pass...")
batch = next(iter(dataloader))
logits_ab, logits_ac, supervised_logits = model(**batch['inputs'])

print("\nOutput shapes:")
print(f"Mol-Morph logits shape: {logits_ab.shape}")
print(f"Mol-Genomic logits shape: {logits_ac.shape}")
print(f"Supervised logits shape: {supervised_logits.shape}")

# Test training step
print("\nTesting training step...")
train_batch_dict = model.training_step(batch, 0)
val_batch_dict = model.validation_step(batch, 0)
```

```python
def compute_logits_comparison(emb_a, emb_c):
    """Compare loop-based and serialized logits computation."""
    batch_size = emb_a.size(0)
    n_cell_lines = emb_c.size(1)
    
    # Current approach (loop-based)
    logits_ac_loop = []
    for c in range(n_cell_lines):
        cell_emb = emb_c[:, c]  # [B, F]
        cell_logits = torch.matmul(emb_a, cell_emb.T)  # [B, B]
        logits_ac_loop.append(cell_logits)
    
    # Proposed approach (serialized)
    # Reshape embeddings for batch matrix multiplication
    emb_a_expanded = emb_a.unsqueeze(1)  # [B, 1, F]
    emb_c_permuted = emb_c.permute(1, 0, 2)  # [C, B, F]
    
    # Compute all logits at once
    logits_ac_serial = torch.matmul(emb_a_expanded, emb_c_permuted.transpose(-1, -2))  # [B, C, B]
    logits_ac_serial = [logits_ac_serial[:, i, :] for i in range(n_cell_lines)]
    
    # Print shapes for verification
    print("\nShape comparison:")
    print(f"emb_a: {emb_a.shape}")  # [B, F]
    print(f"emb_c: {emb_c.shape}")  # [B, C, F]
    print(f"emb_a_expanded: {emb_a_expanded.shape}")  # [B, 1, F]
    print(f"emb_c_permuted: {emb_c_permuted.shape}")  # [C, B, F]
    
    # Verify results are the same
    are_equal = all(
        torch.allclose(loop_logits, serial_logits, atol=1e-6)
        for loop_logits, serial_logits in zip(logits_ac_loop, logits_ac_serial)
    )
    
    print("\nResults comparison:")
    print(f"Shapes match: {all(l1.shape == l2.shape for l1, l2 in zip(logits_ac_loop, logits_ac_serial))}")
    print(f"Values match: {are_equal}")
    
    if not are_equal:
        print("\nSample difference:")
        for i, (l1, l2) in enumerate(zip(logits_ac_loop, logits_ac_serial)):
            if not torch.allclose(l1, l2, atol=1e-6):
                print(f"Cell line {i}:")
                print(f"Max difference: {(l1 - l2).abs().max().item()}")
                break
    
    return logits_ac_loop, logits_ac_serial

# In forward method, you can test both approaches:
logits_ac_loop, logits_ac_serial = compute_logits_comparison(emb_a, emb_c)
# Use either one:
logits_ac = logits_ac_serial  # or logits_ac_loop
```

```python
import pandas as pd
import numpy as np
import os

def split_and_save_genomic_smiles(genomic_data_path, jump_data_path, output_dir, seed):
    # Load genomic data (LINCS)
    genomic_df = pd.read_parquet(genomic_data_path)
    
    # Get unique SMILES from LINCS
    unique_smiles = genomic_df['Metadata_SMILES'].unique()
    total_smiles = len(unique_smiles)
    
    # Calculate split sizes for LINCS
    train_size = int(0.93 * total_smiles)
    
    # Randomly shuffle and split LINCS data with given seed
    np.random.seed(seed)
    shuffled_smiles = np.random.permutation(unique_smiles)
    
    train_smiles = shuffled_smiles[:train_size]
    val_smiles = shuffled_smiles[train_size:]
    
    # Create DataFrames for LINCS splits
    lincs_train_df = pd.DataFrame({'SMILES': train_smiles})
    lincs_val_df = pd.DataFrame({'SMILES': val_smiles})
    
    # Load JUMP data
    jump_train = pd.read_csv(os.path.join(jump_data_path, f'jump-compound-split-{seed}-train.csv'))
    jump_val = pd.read_csv(os.path.join(jump_data_path, f'jump-compound-split-{seed}-val.csv'))
    
    # Create combined datasets
    combined_train = pd.concat([
        jump_train,
        lincs_train_df
    ]).drop_duplicates(subset=['SMILES'])
    
    combined_val = pd.concat([
        jump_val,
        lincs_val_df
    ]).drop_duplicates(subset=['SMILES'])
    
    # Print statistics
    print(f"\nSplit {seed} Statistics:")
    print("LINCS Data:")
    print(f"- Total unique SMILES: {total_smiles}")
    print(f"- Training SMILES: {len(train_smiles)}")
    print(f"- Validation SMILES: {len(val_smiles)}")
    print("\nJUMP Data:")
    print(f"- Training SMILES: {len(jump_train)}")
    print(f"- Validation SMILES: {len(jump_val)}")
    print("\nCombined Data:")
    print(f"- Training SMILES: {len(combined_train)}")
    print(f"- Validation SMILES: {len(combined_val)}")
    
    # Save LINCS splits
    lincs_train_df.to_csv(os.path.join(output_dir, f"LINCS-compound-split-{seed}-train.csv"), index=False)
    lincs_val_df.to_csv(os.path.join(output_dir, f"LINCS-compound-split-{seed}-val.csv"), index=False)
    
    # Save combined splits
    combined_train.to_csv(os.path.join(output_dir, f"JUMP-LINCS-compound-split-{seed}-train.csv"), index=False)
    combined_val.to_csv(os.path.join(output_dir, f"JUMP-LINCS-compound-split-{seed}-val.csv"), index=False)
    
    # Print save locations
    print(f"\nSaved files for split {seed}:")
    print(f"LINCS train: {os.path.join(output_dir, f'LINCS-compound-split-{seed}-train.csv')}")
    print(f"LINCS val: {os.path.join(output_dir, f'LINCS-compound-split-{seed}-val.csv')}")
    print(f"Combined train: {os.path.join(output_dir, f'JUMP-LINCS-compound-split-{seed}-train.csv')}")
    print(f"Combined val: {os.path.join(output_dir, f'JUMP-LINCS-compound-split-{seed}-val.csv')}")
    
    return train_smiles, val_smiles

# Usage example
genomic_data_path = "/scratch/cs/pml/AI_drug/molecular_representation_learning/LINCS/landmark_cmp_data_U2OS_single_dose_time.parquet"
jump_data_path = "/scratch/work/masooda1/mocop/data/jump_data"  # Directory containing JUMP splits
output_dir = "/scratch/work/masooda1/mocop/data/LINCS"

# Create output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

for seed in [0, 1, 2]:
    print(f"\nGenerating split with seed {seed}")
    train_smiles, val_smiles = split_and_save_genomic_smiles(
        genomic_data_path=genomic_data_path,
        jump_data_path=jump_data_path,
        output_dir=output_dir,
        seed=seed
    )
```

```python
import torch
import torch.nn.functional as F
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def original_approach(mol_embeddings, genomic_embeddings, temperature=1.0):
    """
    Original approach: mol embeddings vs all cell lines as classification
    Args:
        mol_embeddings: [batch_size, dim]
        genomic_embeddings: [batch_size, n_cell_lines, dim]
    """
    # Calculate logits
    logits = torch.einsum('bf,bcf->bc', mol_embeddings, genomic_embeddings)  # [B, C]
    logits = logits * temperature
    
    # Calculate loss
    # Assuming each sample matches with its corresponding cell line index
    cell_labels = torch.zeros(logits.size(0), dtype=torch.long).to(logits.device)
    criterion = nn.CrossEntropyLoss(reduction='none')
    losses = criterion(logits, cell_labels)
    
    return logits, losses.mean()

def proposed_approach(mol_embeddings, genomic_embeddings, temperature=1.0):
    """
    Proposed approach: separate contrastive loss for each cell line
    Args:
        mol_embeddings: [batch_size, dim]
        genomic_embeddings: [batch_size, n_cell_lines, dim]
    """
    batch_size, n_cell_lines, dim = genomic_embeddings.shape
    cell_losses = []
    all_logits = []
    
    for cell_idx in range(n_cell_lines):
        # Get embeddings for current cell line
        cell_emb = genomic_embeddings[:, cell_idx, :]  # [B, D]
        
        # Calculate logits for this cell line
        cell_logits = torch.matmul(mol_embeddings, cell_emb.T) * temperature  # [B, B]
        all_logits.append(cell_logits)
        
        # Calculate contrastive loss
        labels = torch.arange(batch_size).to(mol_embeddings.device)
        criterion = nn.CrossEntropyLoss()
        
        # Forward and backward loss
        loss_fwd = criterion(cell_logits, labels)
        loss_bwd = criterion(cell_logits.T, labels)
        cell_loss = (loss_fwd + loss_bwd) / 2
        cell_losses.append(cell_loss)
    
    return torch.stack(all_logits), torch.stack(cell_losses)

# Set random seed for reproducibility
torch.manual_seed(42)

# Generate random embeddings
batch_size = 16
n_cell_lines = 4
embedding_dim = 64

mol_emb = F.normalize(torch.randn(batch_size, embedding_dim), dim=1)
genomic_emb = F.normalize(torch.randn(batch_size, n_cell_lines, embedding_dim), dim=2)

# Compare approaches
print("\nComparing Original vs Proposed Approach:")
print("-" * 50)

# Original approach
orig_logits, orig_loss = original_approach(mol_emb, genomic_emb, temperature=10.0)
print("\nOriginal Approach:")
print(f"Logits shape: {orig_logits.shape}")
print(f"Total loss: {orig_loss.item():.4f}")

# Proposed approach
prop_logits, prop_losses = proposed_approach(mol_emb, genomic_emb, temperature=10.0)
print("\nProposed Approach:")
print(f"Logits shape: {prop_logits.shape}")
print(f"Loss per cell line: {prop_losses.detach().numpy()}")
print(f"Mean loss: {prop_losses.mean().item():.4f}")

# Visualize logits distributions
plt.figure(figsize=(15, 10))

# Original logits
plt.subplot(2, 2, 1)
sns.heatmap(orig_logits.detach().numpy(), cmap='coolwarm', center=0)
plt.title('Original Approach Logits\n[batch_size, n_cell_lines]')
plt.xlabel('Cell Lines')
plt.ylabel('Batch Samples')

# Proposed logits (showing all cell lines)
for i in range(min(3, n_cell_lines)):
    plt.subplot(2, 2, i+2)
    sns.heatmap(prop_logits[i].detach().numpy(), cmap='coolwarm', center=0)
    plt.title(f'Proposed Approach Logits\nCell Line {i} [batch_size, batch_size]')
    plt.xlabel('Batch Samples')
    plt.ylabel('Batch Samples')

plt.tight_layout()
plt.show()

# Print detailed analysis
print("\nDetailed Analysis:")
print("-" * 50)

# Original approach probabilities
orig_probs = F.softmax(orig_logits, dim=1)
print("\nOriginal Approach:")
print("Cell line prediction probabilities:")
print(orig_probs[0].detach().numpy())  # Show first sample
print(f"Max probability: {orig_probs.max(dim=1)[0].mean():.4f}")

# Proposed approach probabilities
print("\nProposed Approach:")
for cell_idx in range(n_cell_lines):
    cell_logits = prop_logits[cell_idx]
    cell_probs = F.softmax(cell_logits, dim=1)
    diag_mean = torch.diagonal(cell_probs).mean()
    print(f"Cell line {cell_idx} - Mean diagonal probability: {diag_mean:.4f}")
```

# Energy function 

```python
import numpy as np
from scipy.spatial.distance import cdist

def compute_energy_distance(X, Y):
    X = np.array(X)
    Y = np.array(Y)
    n1 = X.shape[0]
    n2 = Y.shape[0]

    d_xy = cdist(X, Y, metric='euclidean').mean()
    d_xx = cdist(X, X, metric='euclidean').mean()
    d_yy = cdist(Y, Y, metric='euclidean').mean()

    return 2 * d_xy - d_xx - d_yy
```
